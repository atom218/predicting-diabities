```python
# We start by importing our modules.
import numpy as np
from tensorflow import keras
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.layers import Dense
from keras import Sequential
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction', 'Age']],df.Outcome,test_size=0.2, random_state=25)
```

```python
#similar steps from prediction using sinle neural network
X_train_scaled = X_train.copy()
X_train_scaled = StandardScaler().fit_transform(X_train_scaled)

X_test_scaled = X_test.copy()
X_test_scaled = StandardScaler().fit_transform(X_test_scaled)
```

```python
model=Sequential()
model.add(Dense(16,input_dim=8,activation='relu'))#hidden layer 1 with 16 neurons   
model.add(Dense(8,activation='relu'))#hidden layer 2 with 8 neurons
model.add(Dense(1,activation='sigmoid'))#output layer with single neuron since there is only 1 output

#adam is one of the most effective optimiser for neural network with low memory requirements
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
#epochs is number of iterations as seen in the output
model.fit(X_train,y_train, epochs=500)
```

```python
model.evaluate(X_test_scaled,y_test)
#accuracy is fairly low at 66% 
#there is over fit since accuracy of test data is higher at 78%
```
